This project works with a dataset of used cars from the German version of eBay, to practice with data cleaning and analysis using Python and Pandas.

```python
import numpy as np
import pandas as pd 

autos = pd.read_csv('autos.csv', encoding="Latin-1")
```

```python
autos.info()
autos.head()
```

Looking at the first few rows of the data, the 'name' column seems to require some cleaning, as the formatting is different in each row and the entries are difficult to read. The rest is easy enough to understand for now. 

```python
columns = ["date_crawled", "name", "seller", "offer_type", "price", "abtest", "vehicle_type", "year_of_registration", "gearbox", "power_PS", "model", "odometer", "month_of_registration", "fuel_type", "brand", "not_repaired_damage", "date_created", "nr_of_pictures", "postal_code", "last_seen"]
```

```python
autos.columns = columns 
autos.head()
```

Decided to change the names of each column to snakecase instead of camelcase, to better fit the conventions of Python 

```python
autos.describe(include='all')
autos["odometer"] = autos["odometer"].str.replace("km", "")
autos["odometer"] = autos["odometer"].str.replace(",", "").astype(float)
autos["price"] = autos["price"].str.replace("$", "")
autos["price"] = autos["price"].str.replace(",", "").astype(float)
```

Price and odometer are columns that should be numeric values, but are being stored as text. Converted these columns to be numeric. The column 'nr_of_pictures' has 0 as its value for every single entry, so will keep this in mind and may drop it from the data later. 

```python
autos.head()
```

```python
print(autos["price"].unique().shape)
print(autos["price"].describe())
print(autos["price"].value_counts().head())
```

```python
brands_top = autos["brand"].value_counts().head()
mean_prices = {}
mean_mileage = {}
brand_names = brands_top.index 
for name in brand_names: 
    avg_price = autos.loc[autos["brand"] == name, "price"].mean()
    mean_prices[name] = round(avg_price, 2)
    avg_mileage = autos.loc[autos["brand"] == name, "odometer"].mean()
    mean_mileage[name] = round(avg_mileage, 2)
print(mean_prices)
print(mean_mileage)
```

Looking at the top five most commonly sold auto brands out of this data, I've calculated the average price for each brand to get an idea of what the most common brands are and the relative cost of each. Mercedes-Benz seems to be the most expensive by far, and Opel seems to be the cheapest. 

```python
prices_series = pd.Series(mean_prices)
mileage_series = pd.Series(mean_mileage)
price_vs_mileage = pd.DataFrame(prices_series, columns=["Mean Price"])
price_vs_mileage["Mean Mileage (km)"] = mileage_series
price_vs_mileage
```
